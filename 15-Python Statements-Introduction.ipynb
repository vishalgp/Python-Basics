## What are statements

- At their base, programs written in the Python language are composed of statements and expressions.
- statements are the instructions you write to tell Python what your programs should do.
- Expressions process objects and are embedded in statements.
- Statements code the larger logic of a program’s operation—they use and direct expressions to process the objects we studied in the preceding lectures.

```python
# Following is a expression

2+3-34+23
```

```python
# Following is a statement

print(2+3)
```
